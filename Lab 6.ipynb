<h2>2). Написати функцію, що визначає найкоротше слово в реченні.</h2>

```python
import re
s = ("Місяць - єдиний природний супутник Землі. Найближчий до Сонця супутник планети, так як у найближчих до Сонця планет їх немає. Другою за яскравістю об'єкт на земному небосхилі після Сонця і п'ятий за величиною природний супутник планети Сонячної системи.")
words = re.sub("[^\w]", " ",  s).split()
t = [v for v in words if len(v) == len(min(words))]
from itertools import groupby
t1 = {k:list(v) for k,v in groupby(sorted(words, key=len), len)}
print(t1[min(t1)], t)
```

<h2>3). Написати функцію, що повертає список всіх найкоротших слів в речення.</h2>

```python
string = input("Введіть рядок:")
word = 0
y = 0
for i in range(len(string)):
    if string[i] == "\'":
        string[i] = ' '
for i in string.split():
    word += 1
    if len(i) <= 3:
        y += 1
print("Кількість слів у рядку:")
print(word)
print(y)
```

<h2>5). Написати функцію, що визначає в реченні найдовше слово, що
починається з великої літери. </h2>

```python
import re
s = ("Місяць - єдиний природний супутник Землі. Найближчий до Сонця супутник планети, так як у найближчих до Сонця планет їх немає. Другою за яскравістю об'єкт на земному небосхилі після Сонця і п'ятий за величиною природний супутник планети Сонячної системи.")
words = re.sub("[^\w]", " ",  s).split()
t = [v for v in words if len(v) == len(max(words))]
from itertools import groupby
t1 = {k:list(v) for k,v in groupby(sorted(words, key=len), len)}
print(t1[max(t1)], t)
```

<h2>Проект 2 Написати скрипт для кодування та декодування тексту кодом Цезаря</h2>

```python
llst = ['а','б','в','г','д','е','ж','з','и','й','к','л','м','н','о','п','р','с','т','у','ф','х','ц','ч','ш','щ','ъ','ы','ь','э','ю','я']
blst = ['А','Б','В','Г','Д','Е','Ж','З','И','Й','К','Л','М','Н','О','П','Р','С','Т','У','Ф','Х','Ц','Ч','Ш','Щ','Ъ','Ы','Ь','Э','Ю','Я']
 
 
def encryptCaesar(msg, shift=3):
    ret = ""
    for x in msg:
        if x in llst:
            ind = llst.index(x)
            ret += llst[ind+shift]
        elif x in blst:
            ind = blst.index(x)
            ret += blst[ind+shift]
        else:
            ret += x
    return ret
 
def decryptCaesar(msg, shift=3):
    ret = ""
    for x in msg:
        if x in llst:
            ind = llst.index(x)
            ret += llst[ind-shift]
        elif x in blst:
            ind = blst.index(x)
            ret += blst[ind-shift]
        else:
            ret += x
    return ret
 
print(encryptCaesar("Да здравствует салат Цезарь!"))
print(decryptCaesar("Зг кзугефхецих фгогх Щикгуя!"))
```
